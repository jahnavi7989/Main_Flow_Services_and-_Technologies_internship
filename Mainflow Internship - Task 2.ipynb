```python
data2.drop(['Country', 'Density\n(P/Km2)', 'Abbreviation', 'Agricultural Land( %)','Land Area(Km2)', 'Armed Forces size','Capital/Major City', 'Co2-Emissions', 'CPI', 'CPI Change (%)', 'Currency-Code', 'Forested Area (%)','Gasoline Price', 'GDP', 'Gross primary education enrollment (%)','Gross tertiary education enrollment (%)','Largest city', 'Minimum wage', 'Official language', 'Out of pocket health expenditure', 'Population','Population: Labor force participation (%)', 'Tax revenue (%)','Total tax rate', 'Unemployment rate', 'Urban_population'],axis=1,inplace=True)
data2.columns
```

```python
Q1=data2.quantile(0.25)
Q3=data2.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
data2=data2[~((data2<(Q1-1.5*IQR))|(data2>(Q3+1.5*IQR))).any(axis=1)]
data2
```
